```python
import pandas as pd
data=pd.read_csv("predictions_enn.csv",header=None)
predictions_enn=data.values.tolist()

data=pd.read_csv("lynx_predictions_ann.csv",header=None)
predictions_ann=data.values.tolist()

data=pd.read_csv("lynx_test_Y.csv",header=None)
real=data.values.tolist()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,9))
plt.plot(predictions_enn, label = 'enn', color = 'b')
plt.plot(predictions_ann, label = 'ann', color = 'r')
plt.plot(real, label = 'TEST', color = 'y')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error as mse

error = mse(real, predictions_enn)
print('Test MSE ENN: %.3f' % error)
error = mse(real, predictions_ann)
print('Test MSE ANN: %.3f' % error)
```
