```python
import pandas as pd
data=pd.read_csv("predictions_enn.csv",header=None)
predictions_enn=data.values.tolist()

data=pd.read_csv("predictions_ann.csv",header=None)
predictions_ann=data.values.tolist()
```

```python
import numpy as np
data=pd.read_csv("new_train_Y.csv",header=None)
train_y = data.values.tolist()
data=pd.read_csv("new_train_X.csv",header=None)

train_y = [[float(i)] for i in np.array(data[0:1])[0].tolist()[0].split(' ')] + train_y
```

```python
import matplotlib.pyplot as plt

len_train = 203

plt.figure(figsize=(16,8))
plt.plot(range(len_train), train_y, label = 'train')
plt.plot(range(len_train,len_train + 106),predictions_enn, label = 'enn', color = 'b')
plt.plot(range(len_train,len_train + 106),predictions_ann, label = 'ann', color = 'r')
#plt.plot(range(len_train,len_train + 106),real, label = 'TEST', color = 'y')
plt.legend()
plt.show()
```
